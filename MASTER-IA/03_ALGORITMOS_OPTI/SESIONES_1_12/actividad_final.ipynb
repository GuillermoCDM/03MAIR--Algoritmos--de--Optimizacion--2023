```python
import numpy as np
import random
import copy
import pandas as pd

# definimos una jornada de ejemplo de diapositivas
jornada = ['a_b', 'a_b', 'c_c', 'a_b', 'c_b', 
           'c_c', 'b_b', 'b_b', 'b_b', 'c_b']

# CUANTIFICADORES DEL HORARIO
horario = {'v20' : 0.4 , 's12' : 0.55 , 's16' : 0.7 , 's18' : 0.8 , 's20' : 1 ,
          'd12' : 0.45 , 'd16' : 0.75 , 'd18' : 0.85 , 'd20' : 1 , 'l20' : 0.4}
horario_auxiliar = list(horario.keys())

# CALENDARIO DE LA JORNADA
calendario = {'v20' : np.zeros(10) , 's12' : np.zeros(10) , 's16' : np.zeros(10), 's18' :np.zeros(10) , 's20' :np.zeros(10)  ,
          'd12' :np.zeros(10) , 'd16' :np.zeros(10) , 'd18' :np.zeros(10) , 'd20' : np.zeros(10) , 'l20' :np.zeros(10) }

calendario_0 = pd.DataFrame(calendario, index = jornada)

# A  CADA PARTDO LE APLCIA SU BASE SEGUN LO EQUIPO QUE SE ENFRNETAN
def base_partido(partido):
    if partido == 'a_a':
        return 2
    elif  partido == 'a_b' or  partido == 'b_a':
        return 1.3
    elif partido == 'a_c' or  partido == 'c_a':
        return 1
    elif partido == 'b_b':
        return 0.9
    elif partido == 'b_c' or  partido == 'c_b':
        return 0.75
    elif partido == 'c_c':
        return 0.47
    else : 
        0

def base_horario(partidos, horarios):
    return(base_partido(partidos) * horario[horarios])
        

def base_coincidencia(numero, coincidencias):
    if coincidencias == 0:
        return numero * 1
    elif coincidencias == 1:
        return numero * 0.75
    elif coincidencias == 2:
        return numero * 0.55
    elif coincidencias == 3:
        return numero * 0.4
    elif coincidencias == 4:
        return numero * 0.3
    elif coincidencias == 5:
        return numero * 0.25
    elif coincidencias == 6:
        return numero * 0.22
    elif coincidencias == 7:
        return numero * 0.2
    elif coincidencias == 8:
        return numero * 0.2
    else :
        return 0
    

display(calendario_0)
```

```python
jornada_optima = 0
calendario_optimo = calendario_0
for a in range(10): 
    for b in range(10):
        for c in range(10):
            for d in range(10):
                for e in range(10):
                    for f in range(10):
                        for g in range(10):
                            for h in range(10):
                                for i in range(10):
                                    for j in range(10):
                                        #hago todas las jornadas y comparos sus audiencias y me qquedo con la mejor
                                        cal = pd.DataFrame(calendario, index = jornada)
                                        cal.iloc[0, a] = 1
                                        cal.iloc[1, b] = 1
                                        cal.iloc[2, c] = 1
                                        cal.iloc[3, d] = 1
                                        cal.iloc[4, e] = 1
                                        cal.iloc[5, f] = 1
                                        cal.iloc[6, g] = 1
                                        cal.iloc[7, h] = 1
                                        cal.iloc[8, i] = 1
                                        cal.iloc[9, j] = 1
                                        
                                        mi_max = 0
                                        mi_max = base_coincidencia(base_horario(jornada[0], horario_auxiliar[a]), cal.sum(axis = 0)[0] - 1) +base_coincidencia(base_horario(jornada[1], horario_auxiliar[b]), cal.sum(axis = 0)[1] - 1) + base_coincidencia(base_horario(jornada[2], horario_auxiliar[c]), cal.sum(axis = 0)[2] - 1) + base_coincidencia(base_horario(jornada[3], horario_auxiliar[d]), cal.sum(axis = 0)[3] - 1) +base_coincidencia(base_horario(jornada[4], horario_auxiliar[e]), cal.sum(axis = 0)[4] - 1) + base_coincidencia(base_horario(jornada[5], horario_auxiliar[f]), cal.sum(axis = 0)[5] - 1) + base_coincidencia(base_horario(jornada[6], horario_auxiliar[g]), cal.sum(axis = 0)[6] - 1) + base_coincidencia(base_horario(jornada[7], horario_auxiliar[h]), cal.sum(axis = 0)[7] - 1) +base_coincidencia(base_horario(jornada[8], horario_auxiliar[i]), cal.sum(axis = 0)[8] - 1) + base_coincidencia(base_horario(jornada[9], horario_auxiliar[j]), cal.sum(axis = 0)[9] - 1)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                      
                                            
                                        if mi_max > jornada_optima:      
                                            jornada_optima = mi_max
                                            calendario_optimo = cal.copy()
                                            
                                            
print(f'el caledario optimo es este con numeor maximo de audiencia de :  {jornada_optima}')
display(calendario_optimo)     
```
